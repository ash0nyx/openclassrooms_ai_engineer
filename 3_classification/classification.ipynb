## 1. Data Exploration

```python
#exploration
import pandas as pd
import numpy as np

#titles for graphs
from IPython.display import display, Markdown

#visualisation
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px

import glob
```

```python
# List all CSV files matching the pattern
csvs = glob.glob("extrait_*.csv")
```

```python
def section(text, level=3):
    display(Markdown(f"{'#' * level} {text}"))

def eda_file(df, name):

    #Full EDA for a single DataFrame.

    # Title of the DF
    section(name.capitalize(), level=2)

    #let's get basic info about the data
    section("Overview")
    pd.set_option('display.max_columns', None)
    pd.set_option('display.max_colwidth', None)
    display(df)        # first and last 5 rows - get a feel for the data

    section("Info")
    df.info()                 # column types + non-null counts

    section("Nulls per Column")
    nulls = df.isnull().sum()
    display(nulls[nulls > 0] if nulls.any() else "✅ No nulls")      # null count per column


    section("Duplicate Rows")
    n_dup = df.duplicated().sum()         # number of duplicate rows
    display(f"{n_dup} duplicate row(s)")

    section("Distinct Values per Column")
    display(df.nunique())     # distinct values per column

    section("Constant or Near-Constant Columns (< 2 distinct values)")
    low_card = df.columns[df.nunique() <= 1].tolist()
    display(low_card if low_card else "None")

    #split cols by type
    num_cols  = df.select_dtypes(include='number').columns.tolist()
    cat_cols  = df.select_dtypes(exclude='number').columns.tolist()
    cat_cols = [col for col in cat_cols if col != 'eval_number']

    section("Statistics")
    display(df.describe())    # stats for numeric columns (mean, min, max...)

    section("Histogramms")
    fig, axes = plt.subplots((len(num_cols) + 2) // 3, 3,
                            figsize=(15, 4 * ((len(num_cols) + 2) // 3)))
    axes = axes.flatten()
    for i, col in enumerate(num_cols):
        sns.histplot(df[col], kde=True, ax=axes[i])
        axes[i].set_title(col, fontsize=10)
    for j in range(i + 1, len(axes)):
        axes[j].set_visible(False)
    plt.tight_layout()
    plt.show()

    section("Numeric Columns — Boxplots")
    fig, axes = plt.subplots((len(num_cols) + 2) // 3, 3,
                            figsize=(15, 4 * ((len(num_cols) + 2) // 3)))
    axes = axes.flatten()
    for i, col in enumerate(num_cols):
        sns.boxplot(x=df[col], ax=axes[i])
        axes[i].set_title(col, fontsize=10)
    for j in range(i + 1, len(axes)):
        axes[j].set_visible(False)
    plt.tight_layout()
    plt.show()

    section("Correlation Matrix")
    n = len(df.select_dtypes(include='number').columns)
    fig, ax = plt.subplots(figsize=(n, n))
    sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f", ax=ax)
    plt.tight_layout()
    plt.show()

    if cat_cols:
        section("Categorical Columns — Bar & Donut Charts")
        ncols = len(cat_cols)
        fig = make_subplots(rows=2, cols=ncols,
                            specs=[[{'type': 'xy'}] * ncols,
                                [{'type': 'domain'}] * ncols],
                            subplot_titles=cat_cols + [''] * ncols,
                            vertical_spacing=0.15)

        for i, col in enumerate(cat_cols):
            vc = df[col].value_counts(dropna=False)
            pct = vc / vc.sum()
            colors = px.colors.qualitative.Plotly[:len(vc)]

            fig.add_trace(go.Bar(
                x=vc.index.astype(str),
                y=vc.values,
                hovertemplate='%{x}<br>Count: %{y}<br>Percentage: %{customdata:.1%}<extra></extra>',
                customdata=pct.values,
                marker=dict(color=colors),
                showlegend=False
            ), row=1, col=i + 1)

            # Group into top 9 + Other for donut
            if len(vc) > 9:
                top = vc.head(9)
                other = pd.Series({'Other': vc.iloc[9:].sum()})
                vc_donut = pd.concat([top, other])
            else:
                vc_donut = vc
            donut_colors = px.colors.qualitative.Plotly[:len(vc_donut)]

            fig.add_trace(go.Pie(
                labels=vc_donut.index.astype(str),
                values=vc_donut.values,
                hole=0.4,
                textinfo='label+percent',
                hovertemplate='%{label}<br>Count: %{value}<br>Percentage: %{percent}<extra></extra>',
                marker=dict(colors=donut_colors),
                showlegend=False
            ), row=2, col=i + 1)

        fig.update_layout(height=800, width=500 * ncols,
                        title=f"{name} — Categorical Distributions")
        fig.show()

    # Add a separator for clarity
    display(Markdown("------------------"))

for file in csvs:
    name = file.split("_")[1].split(".")[0]
    df   = pd.read_csv(file)
    eda_file(df, name)
```

### 1.2 Join

sirh.id_employee = eval.eval_number (strip "E_") = sondage.code_sondage (Pandas cut off leading "0")

```python
section("Building the Central DataFrame", level=1)

df_sirh    = pd.read_csv("extrait_sirh.csv")
df_eval    = pd.read_csv("extrait_eval.csv")
df_sondage = pd.read_csv("extrait_sondage.csv")

# Clean join keys before merging
df_eval["id_employee"]    = df_eval["eval_number"].str.replace("E_", "", regex=False).astype(int)
df_sondage["id_employee"] = df_sondage["code_sondage"]

df = (
    df_sirh
    .merge(df_eval,    on="id_employee", how="inner")
    .merge(df_sondage, on="id_employee", how="inner")
)

# Clean % column immediately after joining
df["augementation_salaire_precedente"] = (
    df["augementation_salaire_precedente"]
    .str.replace("%", "", regex=False)
    .str.strip()
    .astype(float)
)

# EDA on joined file
eda_file(df, "Joined")
```
